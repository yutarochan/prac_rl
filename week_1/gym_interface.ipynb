# OpenAI Gym (Demo)

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
env = gym.make('MountainCar-v0')
plt.imshow(env.render('rgb_array'))
print('Observation Space: ' + str(env.observation_space))
print('Action Space: ' + str(env.action_space))
```

```python
obs0 = env.reset()
print('Initial Observation Code: ' + str(obs0))
```

```python
print('Taking Action 2 (right)')
new_obs, reward, is_done, _ = env.step(2)

print('New Observation Code: ' + str(new_obs))
print('reward: ' + str(reward))
print('Game Over: ' + str(is_done))
```

```python
# play with it!
TIME_LIMIT = 250
env = gym.wrappers.TimeLimit(gym.envs.classic_control.MountainCarEnv(), max_episode_steps=TIME_LIMIT + 1)

s = env.reset()
actions = {'left': 0, 'stop': 1, 'right': 2}

%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()

def policy(t, s):
    # Implementation of Simple Policy
    if t < 50:
        return actions['left']
    else:
        return actions['right']


for t in range(TIME_LIMIT):
    s, r, done, _ = env.step(policy(t, s))
    
    # Plot Image
    ax.clear()
    ax.imshow(env.render('rgb_array'))
    fig.canvas.draw()
    
    if done:
        print('YASSSSS!')
        break
else:    
    print('Times up buddy! >:)')
```
